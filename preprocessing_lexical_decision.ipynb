```python
import pandas as pd
import numpy as np
```

```python
participants = pd.read_csv('participants.csv')
data = pd.DataFrame()

for participant_id in participants['participant_ID']:
    participant_data = pd.read_csv(f'output/{participant_id}.csv')
    data = data.append(participant_data)
#print(participant_data)
data
```

adding accuracy score

```python
conditions = [
    (data['condition'] == 'rw') & (data['key'] == 'z'),
    (data['condition'] == 'nw') & (data['key'] == 'm'),
    (data['condition'] == 'rw') & (data['key'] == 'm'),
    (data['condition'] == 'nw') & (data['key'] == 'z')
    ]
values = ['1', '1', '0', '0'] # 1 for correct, 0 for incorrect
data['accuracy'] = np.select(conditions, values)
data

```

```python
data.rename(columns={'Unnamed: 0': 'trial_order'}, inplace=True)
data.to_csv('data.csv', index=False)
```

```python
all_data = data.merge(participants, on='participant_ID')
all_data['participant_ID'] = all_data['participant_ID'].astype(object)
all_data.to_csv('all_data.csv', index=False)
all_data
```
